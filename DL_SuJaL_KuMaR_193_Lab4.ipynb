# **1пёЏвѓЈ Import Libraries**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics import accuracy_score
import time
import matplotlib.pyplot as plt
```

# **2пёЏвѓЈ Load Dataset**

```python
vocab_size = 10000
maxlen = 200

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)

x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)
```

# **3пёЏвѓЈ Function to Train Model**

```python
def train_model(model, name):

    model.compile(
        loss='binary_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )

    start = time.time()

    history = model.fit(
        x_train, y_train,
        epochs=3,
        batch_size=64,
        validation_split=0.2,
        verbose=1
    )

    end = time.time()

    loss, acc = model.evaluate(x_test, y_test, verbose=0)

    print("\n", name, "Results")
    print("Accuracy:", acc)
    print("Training Time:", end - start)

    return acc, history
```

# **4пёЏвѓЈ Simple RNN Model**

```python
rnn_model = Sequential([
    tf.keras.layers.Embedding(vocab_size, 64, input_length=maxlen),
    SimpleRNN(64),
    Dense(1, activation='sigmoid')
])

rnn_acc = train_model(rnn_model, "Simple RNN")
```

# **5пёЏвѓЈ LSTM Model**

```python
lstm_model = Sequential([
    tf.keras.layers.Embedding(vocab_size, 64, input_length=maxlen),
    LSTM(64),
    Dense(1, activation='sigmoid')
])

lstm_acc = train_model(lstm_model, "LSTM")
```

# **6пёЏвѓЈ GRU Model**

```python
gru_model = Sequential([
    tf.keras.layers.Embedding(vocab_size, 64, input_length=maxlen),
    GRU(64),
    Dense(1, activation='sigmoid')
])

gru_acc = train_model(gru_model, "GRU")
```

# **7пёЏвѓЈ Compare Results**

```python
print("\nFinal Comparison")

print("RNN Accuracy :", rnn_acc)
print("LSTM Accuracy:", lstm_acc)
print("GRU Accuracy :", gru_acc)
```
